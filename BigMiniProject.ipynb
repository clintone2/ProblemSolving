```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
from scipy import stats
%matplotlib qt
```

```python
rng = np.random.default_rng()

#for comparing different values of k within n
def tie(Bowl,closedLoops):
    #choose two elements of bowl
    draw = []
    for m in range(2):
        loose = False
        while loose == False:
            draw1 = rng.choice(np.arange(0,len(Bowl)))
            draw2 = rng.choice(np.arange(2))
            if Bowl[draw1][draw2] == 0:
                Bowl[draw1][draw2] = 1
                loose = True
                draw.append(draw1)
    #if we have created a closed loop
    if draw[0] == draw[1]:
        closedLoops = closedLoops + 1
    else:
        Bowl[draw1][draw2] = 0 #this becomes longer string
        newBowl = np.delete(Bowl, draw[0], axis=0)[:]
        Bowl = newBowl[:]
    return(Bowl, closedLoops)

n = 3
results = []
for k in range(1,n+1):
    sample = 10000
    result = []
    for i in range(sample):
        Bowl = np.zeros((n,2), dtype=int)
        closedLoops = 0
        for j in range(k):
            Bowl, closedLoops = tie(Bowl, closedLoops)
        result.append(closedLoops)
    results.append(result)
```

```python
#plotting the frequency distribution of number of closed loops after k ties for n pieces of string
print(results)
Min = 1
Max = n
bins = np.arange(Min-1, Max+2)-0.5
colors = ['purple', 'red', 'pink']
prob, ax = plt.subplots(1,1)
prob, a1, a2, = plt.hist(results, bins=bins, density=True, color=colors)
plt.subplots_adjust(bottom = 0.2)
plt.grid('on')
```

```python
MU = np.mean(results, axis=1)
SIG = np.sqrt(np.var(results, axis=0, ddof=1))
print(MU)
```
